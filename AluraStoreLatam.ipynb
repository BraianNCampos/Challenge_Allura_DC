### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()

tienda2.head()
```

#1. Análisis de facturación



```python
tiendas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}
```

```python
for nombre, df in tiendas.items():
    ingreso_total = df["Precio"].sum()
    print(f"{nombre} - Ingreso total: ${ingreso_total:,.2f}")
```

# 2. Ventas por categoría

```python
for nombre, df in tiendas.items():
    print(f"\n{nombre} - Productos vendidos por categoría:")
    print(df["Categoría del Producto"].value_counts())
```

# 3. Calificación promedio de la tienda


```python
for nombre, df in tiendas.items():
    calificacion_promedio = df["Calificación"].mean()
    print(f"{nombre} - Calificación promedio: {calificacion_promedio:.2f}")
```

# 4. Productos más y menos vendidos

```python
for nombre, df in tiendas.items():
    conteo_productos = df["Producto"].value_counts()
    mas_vendido = conteo_productos.idxmax()
    menos_vendido = conteo_productos.idxmin()
    print(f"\n{nombre}:")
    print(f"Producto más vendido: {mas_vendido} ({conteo_productos.max()} ventas)")
    print(f"Producto menos vendido: {menos_vendido} ({conteo_productos.min()} ventas)")
```

# 5. Envío promedio por tienda

```python
for nombre, df in tiendas.items():
    costo_envio_promedio = df["Costo de envío"].mean()
    print(f"{nombre} - Costo de envío promedio: ${costo_envio_promedio:.2f}")
```

# INGRESO POR TIENDA


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ingresos por tienda
ingresos = {nombre: df["Precio"].sum() for nombre, df in tiendas.items()}

plt.figure(figsize=(8, 5))
sns.barplot(x=list(ingresos.keys()), y=list(ingresos.values()), palette="Blues_d")
plt.title("Ingresos Totales por Tienda")
plt.ylabel("Ingresos ($)")
plt.xlabel("Tienda")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Distribución de categorías (TIENDA 1)

```python
categoria_t1 = tienda["Categoría del Producto"].value_counts()

plt.figure(figsize=(10, 5))
sns.lineplot(x=categoria_t1.index, y=categoria_t1.values, marker='o')
plt.title("Productos Vendidos por Categoría - Tienda 1")
plt.ylabel("Cantidad Vendida")
plt.xlabel("Categoría del Producto")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
